```python
# Import pyspark library

from pyspark.sql import SparkSession

# Initiate PySpark Session

spark = SparkSession.builder.appName('PySpark_Guide').getOrCreate()
```

```python
# read csv file

df_pyspark = spark.read.csv('test2.csv', header=True, inferSchema=True)
```

**DATA FRAME**

```python
# Libraries required

from pyspark.sql import Row
from datetime import date, datetime
import pandas as pd

# create pyspark dataframes using list of rows

df_spark = spark.createDataFrame([
    Row(a=1, b=2, c='string1', d=date(2000, 1, 1), e=datetime(2000, 1, 1, 12, 0)),
    Row(a=1, b=4, c='string2', d=date(2000, 1, 3), e=datetime(2000, 1, 1, 11, 0)),
])

df_spark.show()
```

```python
# create pyspark dataframes using explicit schema

df_spark = spark.createDataFrame([
    (1, 2.0, 'string1', date(2000, 1, 1), datetime(2000, 1, 1, 12, 0)),
    (1, 7.0, 'string1', date(2000, 1, 1), datetime(2000, 1, 1, 12, 0)),
    (1, 4.0, 'string1', date(2000, 1, 1), datetime(2000, 2, 1, 13, 0))
], schema = ['a long', 'b double', 'c string', 'd date', 'e timestamp'])

df_spark.show()
```

```python
# create pyspark dataframes using pandas dataframes

df = pd.DataFrame({
    'a' : [1, 2, 3],
    'b' : [2.0, 3.0, 4.0],
    'c' : ['string1', 'string2', 'string3'],
    'd' : [date(2000, 1, 1), date(2000, 3, 4), date(2000, 4, 8)],
    'e' : [datetime(2000, 1, 2, 12, 0), datetime(2000, 9, 2, 12, 0), datetime(2000, 12, 2, 12, 0)]
})

df_spark = spark.createDataFrame(df)
df_spark.show()
```

```python
df_pyspark.show()
df_pyspark.printSchema()
```

```python
# pyspark provides diversity of using serveral user defined functions and API's of python to be integrated with spark

# Libraries

from pyspark.sql.functions import pandas_udf
import pandas as pd

# create a function to add 2 to every row of age column

@pandas_udf('long')
def pandas_plus_two(series : pd.Series) -> pd.Series:
  return series+2

# apply the udf on age column
df_pyspark.select(pandas_plus_two(df_pyspark.Age)).show()
```

```python
# create a pyspark dataframe using list of rows and apply group and avg function to get the desired result

from pyspark.sql import Row

df_spark = spark.createDataFrame([
    Row(a=1, b=2.0, c='string1', d=date(2000, 1, 1), e=datetime(2000, 1, 1, 12, 0)),
    Row(a=2, b=2.0, c='string1', d=date(2000, 1, 1), e=datetime(2000, 1, 1, 12, 0)),
    Row(a=3, b=2.0, c='string1', d=date(2000, 1, 1), e=datetime(2000, 1, 1, 12, 0)),
])

df_spark.show()

df_spark.groupby('a').avg().show()
```
